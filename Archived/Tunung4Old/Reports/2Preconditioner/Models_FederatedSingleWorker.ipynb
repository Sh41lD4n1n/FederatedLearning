## Example 1 Global update on each iterations

```python
def show_data_stat(current_stat):
    #current_stat = federated_sch_u_stat_collectors[0]
    current_stat.loss_plot()
    current_stat.acc_plot()

    current_stat.plot_confusion_matrix()
    current_stat.print_accuracy()

    current_stat.show_y_distr()
    
def show_data_stat_comparison(current_stat_collector1,current_stat_collector2):
    #stat_np,stat_origin
    #current_stat_collector1 = federated_sch_u_stat_collectors[0]
    #current_stat_collector2 = federated_sch_stat_collectors[0]

    current_stat_collector1.comparison_loss_plot(current_stat_collector2)
    current_stat_collector1.comparison_accuracy_plot(current_stat_collector2)

    print(current_stat_collector1.lable)

    current_stat_collector1.plot_confusion_matrix()
    current_stat_collector1.print_accuracy()

    print(current_stat_collector2.lable)
    current_stat_collector2.plot_confusion_matrix()
    current_stat_collector2.print_accuracy()

    current_stat_collector1.comparison_mse(current_stat_collector2)
```

### Federated, preconditioner: SGD

#### Scheduler: identity

```python
#SGD
model_params = {"lr":LR_EPOCH['lr']}

precond = [SGD_preconditioner()]

federated_is_stat_collectors = [Statistic(X_test,y_test,"worker1_ident_sgd")]

models = [Logistic_Regression_numpy(n_features = X_train.shape[1],optimizer = None,
                                       lr = model_params['lr'], preconditioner = precond[0],
                                       epoch = 1,stat_collector=federated_is_stat_collectors[0])]
scheduler1 = [MatrixScheduler_identical()]


s = server(M_workers=1,
           X_train = X_train,y_train = y_train,
           stat_collectors = federated_is_stat_collectors,
           models = models,
           data_splitter = 'id',m_scheduler = scheduler1)

s.run(n_iter=LR_EPOCH['epoch'],T=1)
```

Loss, Weigths, Accuracy Plots

```python
show_data_stat(federated_is_stat_collectors[0])
```

#### Scheduler: uniform

```python
#SGD
model_params = {"lr":LR_EPOCH['lr']}

precond = [SGD_preconditioner()]

federated_us_stat_collectors = [Statistic(X_test,y_test,"worker1_uniform_sgd")]

models = [Logistic_Regression_numpy(n_features = X_train.shape[1],optimizer = None,
                                       lr = model_params['lr'], preconditioner = precond[0],
                                       epoch = 1,stat_collector=federated_us_stat_collectors[0])]
scheduler1 = [MatrixScheduler_Uniform(alpha=0.01,
                                      gamma=3,
                                      change_mode=MatrixScheduler_Uniform.change_on_each_iteration)]


s = server(M_workers=1,
           X_train = X_train,y_train = y_train,
           stat_collectors = federated_us_stat_collectors,
           models = models,
           data_splitter = 'id',m_scheduler = scheduler1)

s.run(n_iter=LR_EPOCH['epoch'],T=1)
```

Loss, Weigths, Accuracy Plots

```python
show_data_stat(federated_us_stat_collectors[0])
```

#### сравнение: Scheduler: identity,Scheduler: uniform

```python
show_data_stat_comparison(current_stat_collector1 = federated_is_stat_collectors[0],
                          current_stat_collector2 = federated_us_stat_collectors[0])
```
