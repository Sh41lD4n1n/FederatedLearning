```python
def k_scheduler_statistic_draw(servers,labels,y_lim):
    
    r = [np.arange(len(servers)) for _ in range(len(servers[0].workers))]
    barWidth = 0.3
    for i in range(1,len(servers[0].workers)):
        r[i] = [x + barWidth for x in r[i-1]]
    
    
    values = [[] for _ in range(len(servers[0].workers))]
    for s in servers:
        for i,w in enumerate(s.workers):
            y_pred = w.test(X_test)
            y_pred_classes = y_pred.copy()

            threshlod = 0.5
            y_pred_classes[y_pred_classes>threshlod] = 1
            y_pred_classes[y_pred_classes<1] = 0

            #bas = balanced_accuracy_score(y_test,y_pred_classes)
            acc = accuracy_score(y_test,y_pred_classes)

            values[i].append(acc*100)
    
    for i in range(len(r)):
        #plt.bar(r[i], values[i], width = barWidth, color = 'blue', edgecolor = 'black', capsize=7, label='poacee')
        plt.bar(r[i], values[i], width = barWidth, edgecolor = 'black', capsize=7, label=f'worker{i}')
        
    plt.ylim(y_lim[0], y_lim[1])
    
    plt.xticks([r + barWidth for r in range(len(labels))], labels)
    plt.ylabel('acc')
    plt.legend()
 
    # Show graphic
    plt.show()
```

```python
def k_scheduler_statistic_draw(servers_total,labels,y_lim):
    
    r = [np.arange(len(servers_total[0])) for _ in range(len(servers_total[0][0].workers))]
    barWidth = 0.3
    for i in range(1,len(servers_total[0][0].workers)):
        r[i] = [x + barWidth for x in r[i-1]]
    
    val = np.zeros((len(servers_total[0][0].workers),len(servers_total[0])))
    
    for servers in servers_total:
        values = [[] for _ in range(len(servers_total[0][0].workers))]
        for s in servers:
            for i,w in enumerate(s.workers):
                y_pred = w.test(X_test)
                y_pred_classes = y_pred.copy()

                threshlod = 0.5
                y_pred_classes[y_pred_classes>threshlod] = 1
                y_pred_classes[y_pred_classes<1] = 0

                #bas = balanced_accuracy_score(y_test,y_pred_classes)
                acc = accuracy_score(y_test,y_pred_classes)

                values[i].append(acc*100)
        val += np.array(values)
    values = (val/len(servers_total)).tolist()
    
    for i in range(len(r)):
        #plt.bar(r[i], values[i], width = barWidth, color = 'blue', edgecolor = 'black', capsize=7, label='poacee')
        plt.bar(r[i], values[i], width = barWidth, edgecolor = 'black', capsize=7, label=f'worker{i}')
        
    plt.ylim(y_lim[0], y_lim[1])
    
    plt.xticks([r + barWidth for r in range(len(labels))], labels)
    plt.ylabel('acc')
    plt.legend()
 
    # Show graphic
    plt.show()
```

## Federated learning Tune

```python
def loss(y_true,y):
    return (-y_true*np.log(y) - (1 - y_true)*np.log(1-y)).sum()
    
def tune_model1(model_name):
    
    lr = [1e-3,1e-5]
    num_com = [i for i in range(1,15,1)]
    T = [i for i in range(1,80,20)]
    
    global X_train,y_train,X_test,y_test
    global y_pred_original_proba
    global w_origin
    
    report = {"lr":[],"num_com":[],"T":[],"worker_num":[],"loss":[],"mse_origin":[],"mse_origin_w":[],"acc":[],"min":[],"max":[],"mean":[],"pred":[],"weights":[]}
    for l in lr:
        for c in num_com:
            for t in T:
                s = server(f_model_name = model_name,model_params = {"lr":l},
                           M_workers=3,
                           X_train = X_train,y_train = y_train,data_splitter='id')
                s.run(n_iter=c*t,T=t)
                for i,w in enumerate(s.workers):
                    
                    y_pred = w.test(X_test)
                    y_pred = np.nan_to_num(y_pred,nan=-1)
                    
                    y_pred_classes= y_pred.copy()
                    y_pred_classes[y_pred_t_classes>y_pred.mean()] = 1
                    y_pred_classes[y_pred_t_classes<1] = 0
                    
                    report["lr"].append(l)
                    report["num_com"].append(c)
                    report["T"].append(t)
                    report["worker_num"].append(i)
                    
                    report["loss"].append(loss(y_test,y_pred))
                    report["weights"].append(w.model.get_weights())
                    report["pred"].append(y_pred.copy())
                    report["mse_origin"].append( ((y_pred_original_proba - y_pred)**2).sum() / y_pred.shape[0] )
                    report["acc"].append( accuracy_score(y_test,y_pred_classes) )
                    report["mse_origin_w"].append( ((w_origin - w.model.get_weights())**2) .sum() )
                    
                    report["min"].append(y_pred.min())
                    report["max"].append(y_pred.max())
                    report["mean"].append(y_pred.mean())
            
    return report
            
```
